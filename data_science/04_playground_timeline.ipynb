```python
events = [
    ('2018-11-19', '報酬隠しで逮捕'),
    ('2018-12-10', '報酬隠しで再逮捕'),
    ('2019-12-21', '特別背任で再逮捕'),
    ('2019-01-11', '追起訴、保釈請求（後に却下）'),
    ('2019-03-05', '3度目の保釈請求で許可'),
    ('2019-03-06', '保釈（逮捕から108日目）'),
    ('2019-04-04', '特別背任容疑で再逮捕')
]
```

```python
import pandas as pd

df = pd.DataFrame(events)
df.columns = ['event_timestamp', 'event']
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
import seaborn as sns

sns.set()
rcParams['font.family'] = 'IPAexGothic'
fig, ax = plt.subplots(figsize=(8,16))



y_pos = np.arange(len(df))
date_width = np.ones(len(df)) * 1
content_width = date_width * 2
ax.barh(y_pos, date_width, color='b')
ax.barh(y_pos, content_width, left= date_width*1.01, color='gray')
ax.invert_yaxis()
ax.set_title("ゴーン容疑者をめぐる動き" , fontsize=28)
event_column = 'event'
event_ts_column = 'event_timestamp'
for index, row in df.iterrows():
    ax.text( 0 , index, row[event_ts_column], verticalalignment='top',horizontalalignment='left' ,fontsize=14, color='white')
    ax.text( 1 , index, row[event_column], verticalalignment='top',horizontalalignment='left' ,fontsize=14)
```
